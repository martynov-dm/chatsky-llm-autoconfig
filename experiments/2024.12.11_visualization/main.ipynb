```python
import json
from chatsky_llm_autoconfig.graph import Graph
from chatsky_llm_autoconfig.algorithms.dialogue_generation import DialoguePathSampler, DialogueSampler
```

```python
with open('complex_graphs.json', encoding="utf-8") as f:
    data = json.load(f)
print(len(data))
```

```python
g = Graph(graph_dict=data[-1]['graph'])
g.visualise()
```

```python
d = DialoguePathSampler()
dials = d.invoke(g)
print(len(dials))
for i in dials:
    print(i)
    print("------------")
```
